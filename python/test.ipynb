#### 3 R retention

```python
first_visit = classroom_df.groupby('user_id')['date'].min().reset_index()
first_visit.columns = ['user_id', 'first_date']
first_visit.sort_values('first_date')
```

```python
retention_df = pd.merge(classroom_df, first_visit, on='user_id')
retention_df['active_days'] = (retention_df.date - retention_df.first_date).dt.days
retention_df
```

```python
retention = retention_df.groupby(['first_date', 'active_days'])['user_id'].nunique().unstack(fill_value=0)
retention
```

```python
retention_per = round(retention.divide(retention[0], axis=0) * 100, 2)
retention_per
```

#### 8 AB-test p-value

```python
ab_df = pd.read_excel('Данные для тестового задания.xlsx', sheet_name='Данные АБ тестов')
print('Dataframe len {}'.format(len(ab_df)))
ab_df
```

```python
# группы тестирования
first_test = ab_df[ab_df['experiment_num'] == 1]
second_test = ab_df[ab_df['experiment_num'] == 2]
third_test = ab_df[ab_df['experiment_num'] == 3]
print(len(first_test), len(second_test), len(third_test))
```

```python
# первая группа
test1 = first_test[first_test['experiment_group'] == 'test']['revenue']
control1 = first_test[first_test['experiment_group'] == 'control']['revenue']

test1_mean = np.mean(test1)
control1_mean = np.mean(control1)

test1_std = np.std(test1, ddof=1)
control1_std = np.std(control1, ddof=1)

test1_size = len(test1)
control1_size = len(control1)

dof1 = test1_size + control1_size - 2

t_stat1 = abs(test1_mean - control1_mean) / np.sqrt(test1_std**2 / test1_size + control1_std**2 / control1_size)
p_value_1 = 1 - t.cdf(t_stat1, dof1)
p_value_1
```

```python
# вторая группа
test2 = second_test[second_test['experiment_group'] == 'test']['revenue']
control2 = second_test[second_test['experiment_group'] == 'control']['revenue']

test2_mean = np.mean(test2)
control2_mean = np.mean(control2)

test2_std = np.std(test2, ddof=1)
control2_std = np.std(control2, ddof=1)

test2_size = len(test2)
control2_size = len(control2)

dof2 = test2_size + control2_size - 2

t_stat2 = abs(test2_mean - control2_mean) / np.sqrt(test2_std**2 / test2_size + control2_std**2 / control2_size)
p_value_2 = 1-t.cdf(t_stat2, dof2)
p_value_2
```

```python
# третяя группа
test3 = third_test[third_test['experiment_group'] == 'test']['revenue']
control3 = third_test[third_test['experiment_group'] == 'control']['revenue']

test3_mean = np.mean(test3)
control3_mean = np.mean(control3)

test3_std = np.std(test3, ddof=1)
control3_std = np.std(control3, ddof=1)

test3_size = len(test3)
control3_size = len(control3)

dof3 = test3_size + control3_size - 2

t_stat3 = abs(test3_mean - control3_mean) / np.sqrt(test3_std**2 / test3_size + control3_std**2 / control3_size)
p_value_3 = 1-t.cdf(t_stat3, dof3)
p_value_3
```
